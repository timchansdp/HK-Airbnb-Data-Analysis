# **Project Title: Hong Kong Airbnb Data Analysis**

# **Section 1: Business Understanding** 

The objective of this project is to analyze the open-source Hong Kong Airbnb data by employing statistics and visualizations so that five business questions can be compellingly answered:

- Q1. Which neighborhood has the greatest potential of maximizing the revenue?
- Q2. Is being a super host worthy?
- Q3. Which month is best for visits in terms of availability and pricing?
- Q4. Which neighborhood you should start looking into first in terms of the variety of choices?
- Q5. What are the impact of the 2019 Hong Kong protests and the COVID-19 pandemic from the Airbnb perspective?

In summary, revenue, availability, and pricing are the main areas of interest in this analysis.

This analytics provides an overview of some aspects of the Hong Kong Airbnb market, and can also serve as a guide to all future hosts and visitors, as it answers questions such as the seasonality of availability, or how the super host status influences the revenue.

**Contexts of the questions:** <br>

Q1.) Some people may want to buy a property for value preservation and are willing to rent it on Airbnb so that they can maximize their profit. This group of hosts-to-be may want to know which neighborhood has the highest earning capability before buying a new property.<br>

Q2.) Hosts may want know to how the super host title affects their revenue before spending extra effort to meet the requirements. <br>

Q3.) Local or foreign tourists may want to know which month when the Airbnb market generally has more available accommodations and lower pricing so that they can better decide the vacation dates. <br>

Q4.) Due to the COVID-19 pandemic, taking a local trip is getting popular in Hong Kong. For some local tourists, the location may not be the top priority. They probably just want to be on their vacation in a place with a novel living environment to escape from their lives for a few days. The variety of choices is more important to them. However, people will choose listings based on their budget. Hence, I try to find out which neighborhood can satisfy most of them. <br>

Q5.) Harming the economy was the common criticism of the 2019 Hong Kong protests and the COVID-19 pandemic. It was generally believed as true. However, a social-minded Hong Konger may want to know how they exactly influenced the economy. The Airbnb market could be one of the perspectives to examine this assertion.

# **Section 4: Result Evaluations** 

### Declared functions

```python
# Remove outliers in each neighbourhood

neighbourhood_list = df_raw_listing['neighbourhood_cleansed'].unique().tolist()

def remove_outlier_neighbourhood(df, \
                                 dependent_var, \
                                 review_boolean, \
                                 neighbourhood_list = neighbourhood_list):
    """
    Remove outliers in data grouped by neighbourhoods
  
    Parameters:
      df (pandas dataframe): the data source and outliers will be removed from it
      dependent_var (str): variable that help determine outliers
      review_boolean (boolean): whether to consider review numbers when determining outliers
      neighbourhood_list (list): the names of neighbourhoods

    Returns:
      df (pandas dataframe): the data source without outliers

    """
    
    for nb in neighbourhood_list:

        df.reset_index(drop = True, inplace = True)

        df_temp = df[ df['neighbourhood_cleansed'] == nb ][dependent_var]

        Q1 = np.percentile(df_temp, 25, interpolation = 'midpoint')
        Q3 = np.percentile(df_temp, 75, interpolation = 'midpoint')

        IQR = Q3 - Q1
        
        if not review_boolean:

            above_upper_bound = np.where( (df[dependent_var] >= (Q3 + 1.5 * IQR)) \
                                          & \
                                          (df['neighbourhood_cleansed'] == nb) )
            
            below_lower_bound = np.where( (df[dependent_var] <= (Q3 - 1.5 * IQR)) \
                                          & \
                                          (df['neighbourhood_cleansed'] == nb) )
            
        else:

            above_upper_bound = np.where( (df[dependent_var] >= (Q3 + 1.5 * IQR)) \
                                          & \
                                          (df['neighbourhood_cleansed'] == nb) \
                                          & \
                                          (df['number_of_reviews'] == 0) )
            
            below_lower_bound = np.where( (df[dependent_var] <= (Q3 - 1.5 * IQR)) \
                                          & \
                                          (df['neighbourhood_cleansed'] == nb) \
                                          & \
                                          (df['number_of_reviews'] == 0) )            

        df.drop(above_upper_bound[0], inplace = True)
        df.drop(below_lower_bound[0], inplace = True)

    return df
```

### Question 1: Which neighborhood has the greatest potential of maximizing the revenue?

**Analysis:** <br>
The crucial part is how to measure the earning capability of the neighborhood for comparison. Total revenue within a certain duration could be a suitable indicator. Unfortunately, revenues are not directly provided by the dataset. To estimate the revenues, we have to assume that the booking number of each listing is proportional to the review number. <br>

Then, the revenue of each listing can be estimated as follow:
[price] x [minimum nights] x [review number] <br>

The final step is aggregating the revenues by neighborhoods and then checking out the ranking to determine the answer.

```python
# remove listings that have not been reviewed since 2014-06-05
df_cal_Q1_listing = df_Q1_listing.loc[df_Q1_listing.last_review > '2014-06-05', :].copy()
```

```python
# find listings' number of review given after 2014-06-05
df_review_num_after_2014_06_05 = \
df_Q1_review.where(df_Q1_review.date >= '2014-06-05').groupby(['listing_id']).count()

df_review_num_after_2014_06_05.rename(columns={"date": "review_num_after_2014_06_05"}, inplace = True)

df_cal_Q1_listing = \
df_cal_Q1_listing.merge(df_review_num_after_2014_06_05, how = 'inner', on = 'listing_id')
```

```python
# remove listings with abnormal review number
df_cal_Q1_listing['max_possible_review_num'] = \
( (df_cal_Q1_listing.loc[:, 'last_review'] \
   - df_cal_Q1_listing.loc[:, 'first_review']) \
   / df_cal_Q1_listing.loc[:, 'minimum_nights'] ).dt.days

df_cal_Q1_listing['difference'] = \
df_cal_Q1_listing.loc[:, 'max_possible_review_num'] \
- df_cal_Q1_listing.loc[:, 'number_of_reviews']

df_cal_Q1_listing = \
df_cal_Q1_listing[ (df_cal_Q1_listing['difference'] >= 0) | (df_cal_Q1_listing['number_of_reviews'] == 1)].copy()
```

```python
# calculate listings' estimated revenue of each booking
df_cal_Q1_listing['estimated_revenue'] = \
df_cal_Q1_listing.loc[:, 'price'] * df_cal_Q1_listing.loc[:, 'minimum_nights']


## calculate listings' estimated total revenue from 2014-06-05 to 2021-09-24
df_cal_Q1_listing['total_revenue'] = \
df_cal_Q1_listing.loc[:, 'estimated_revenue'] * df_cal_Q1_listing.loc[:, 'review_num_after_2014_06_05']


# prepare listing df for Q5
df_Q5_listing = df_cal_Q1_listing.copy()
```

```python
# Remove outliers in each neighbourhood
df_cal_Q1_listing = remove_outlier_neighbourhood(df = df_cal_Q1_listing, \
                                                 dependent_var = 'total_revenue', \
                                                 review_boolean = False)
```

```python
# obtain the average total revenues of neighborhoods during 2014-06-05 - 2021-09-24
df_average_revenue_of_neighbourhood = \
df_cal_Q1_listing[['neighbourhood_cleansed', 'total_revenue']].groupby(['neighbourhood_cleansed'])\
                                                              .mean()\
                                                              .sort_values('total_revenue', ascending = False)\
                                                              .astype(int)
```

### Question 2: Is being a super host worthy?

**Analysis**: <br>
The main concern of the question is whether the listings managed by super hosts have higher revenue. The average annual revenue could be an appropriate indicator for this case. After finishing the calculation, we just need to divide the data into the super host group and the non-super host group so that the distributions of them can be plotted. The group generally being distributed on the right of another one has the higher earning capability. 99% Confidence interval of the difference in medians could be used if the distributions are not distinct enough to give out a conclusion.

```python
# have a look at df_Q2_listing
df_Q2_listing
```

##### calculate average annual revenue

```python
# get the df for calculation
df_cal_Q2_listing = df_Q2_listing.copy()
```

```python
# Remove listings with abnormal review number
df_cal_Q2_listing['max_possible_review_num'] = \
( (df_cal_Q2_listing.loc[:, 'last_review'] \
   - df_cal_Q2_listing.loc[:, 'first_review']) \
   / df_cal_Q2_listing.loc[:, 'minimum_nights'] ).dt.days

df_cal_Q2_listing['difference'] = \
df_cal_Q2_listing.loc[:, 'max_possible_review_num'] \
- df_cal_Q2_listing.loc[:, 'number_of_reviews']

df_cal_Q2_listing = \
df_cal_Q2_listing[ (df_cal_Q2_listing['difference'] >= 0) | (df_cal_Q2_listing['number_of_reviews'] == 1)].copy()
```

```python
# Calculate listings' estimated revenue of each booking
df_cal_Q2_listing['estimated_revenue'] = \
df_cal_Q2_listing.loc[:, 'price'] * df_cal_Q2_listing.loc[:, 'minimum_nights']


# Calculate listings' estimated total revenue
df_cal_Q2_listing['total_revenue'] = \
df_cal_Q2_listing.loc[:, 'estimated_revenue'] * df_cal_Q2_listing.loc[:, 'number_of_reviews']


# Calculate listings' estimated average annual revenue
end_date = np.datetime64('2021-09-24')
df_cal_Q2_listing['avg_revenue_per_year'] = \
df_cal_Q2_listing.loc[:, 'total_revenue'] / ((end_date - df_cal_Q2_listing.loc[:, 'host_since']) / np.timedelta64(1, 'Y'))
df_cal_Q2_listing['avg_revenue_per_year'] = df_cal_Q2_listing['avg_revenue_per_year'].astype(int)
```

#### Q2 results

```python
# Seperate the dataframe, listing_df_Q2_cal, into superhost and non-superhost dataframes
df_temp = df_cal_Q2_listing[['host_is_superhost', 'avg_revenue_per_year']].set_index('host_is_superhost').stack()
df_false = df_temp.xs(False, level = ("host_is_superhost")).reset_index(drop = True).copy()
df_true = df_temp.xs(True, level = ("host_is_superhost")).reset_index(drop = True).copy()


# Remove outliers in df_true & df_false
for df in (df_true, df_false):

    Q1 = np.percentile(df, 25, interpolation = 'midpoint')
    Q3 = np.percentile(df, 75, interpolation = 'midpoint')

    IQR = Q3 - Q1

    above_upper_bound = np.where( df >= (Q3 + 1.5 * IQR) )
    below_lower_bound = np.where( df <= (Q3 - 1.5 * IQR) )

    df.drop(above_upper_bound[0], inplace = True)
    df.drop(below_lower_bound[0], inplace = True)

    df.reset_index(drop = True, inplace = True)
```

Below are the histograms and the kernel density estimate (KDE) plots that show how the average annual revenues of the super host group, and the non-super host group were distributed.

```python
#Plot the distributions of superhost and non superhost

# Set plot preference
plt.style.use(style = 'default')

plt.figure(figsize = [20, 5])

# main plots
bin_size = 0.1
df_true_log = np.log10(df_true)
max_value = round(df_true_log.max(), 1)
bins = 10 ** np.arange(0, max_value + bin_size, bin_size)
sns.histplot(data = df_true, \
             bins = bins, \
             stat = 'density', \
             kde = True, \
             color = 'red')
sns.histplot(data = df_false, \
             bins = bins, \
             stat = 'density', \
             kde = True)

plt.legend(['superhost','non superhost'], fontsize = 'x-large')

# set limit for x-asis & scale the x-axis to log
plt.xscale('log')
plt.xlim([10 ** 1, 10 ** 4])

# ticks labeling
ticks = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000]
labels = ['{}'.format(v) for v in ticks]
plt.xticks(ticks, labels)

# ticks formatting
tick_fontsize = 14
plt.xticks(fontsize = tick_fontsize)
plt.yticks(fontsize = tick_fontsize)

# labels & title formatting
label_fontdict = \
{'fontsize': 16, \
 'fontweight' : 'semibold'}
labelpad = 20
pad = 40
plt.xlabel('avg. annual revenue (HK$)', \
           fontdict = label_fontdict, \
           labelpad = labelpad)
plt.ylabel('density', \
           fontdict = label_fontdict, \
           labelpad = labelpad)
plt.title('Average Annual Revenue Distribution of Superhost and Non-superhost', \
          fontdict = label_fontdict, \
          pad = pad);
```

**Explanation:** <br>

In general, we can see that the super host group was mainly distributed on the right of the non-super host group. Given that the visualization may not be evident enough to claim that the median of the super host group was greater, it was safer to give out a conclusion by calculating the 99% confidence interval of the difference in medians of the super host group and the non-super host group.

```python
# obtain the confidence interval for medians of superhost and non-superhost 
# and difference of medians of them by bootstraping

median_list_super = []
median_list_non_super = []
median_diff_list = []

for _ in range(10000):
    
    sample_super = df_true.sample(200, replace = True)
    sample_non_super = df_false.sample(200, replace = True)

    median_super = sample_super.median()
    median_non_super = sample_non_super.median()

    median_list_super.append(median_super)
    median_list_non_super.append(median_non_super)
    median_diff_list.append(median_super - median_non_super)
```

```python
# plot the distribution with 99% confidence interval

# Set plot preference
plt.style.use(style = 'default')

plt.figure(figsize = [20, 5])

# main plots
bin_size = 0.1
df_true_log = np.log10(df_true)
max_value = round(df_true_log.max(), 1)
bins = 10 ** np.arange(0, max_value + bin_size, bin_size)
sns.histplot(data = df_true, \
             bins = bins, \
             stat = 'density', \
             kde = True, \
             color = 'red')
sns.histplot(data = df_false, \
             bins = bins, \
             stat = 'density', \
             kde = True)

# Indicate the difference in median of 2 groups with 99% confidence interval
sample_median_diff = df_true.median() - df_false.median()
margin_error = \
(np.percentile(median_diff_list, 99.5) - np.percentile(median_diff_list, 0.5)) / 2

output_string = f"Difference in median: HK${sample_median_diff: n} (+/-{margin_error: n})"
props = dict(boxstyle = 'round', facecolor = 'white', alpha = 1)
plt.text(1000, 0.0008, \
         output_string , fontsize = 18, \
         verticalalignment = 'bottom', bbox = props)

plt.legend(['superhost','non superhost'], fontsize = 'x-large')

# set limit for x-asis & scale the x-axis to log
plt.xscale('log')
plt.xlim([10 ** 1, 10 ** 4])

# ticks labeling
ticks = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000]
labels = ['{}'.format(v) for v in ticks]
plt.xticks(ticks, labels)

# ticks formatting
tick_fontsize = 14
plt.xticks(fontsize = tick_fontsize)
plt.yticks(fontsize = tick_fontsize)

# labels & title formatting
label_fontdict = \
{'fontsize': 16, \
 'fontweight' : 'semibold'}
labelpad = 20
pad = 40
plt.xlabel('avg. annual revenue (HK$)', \
           fontdict = label_fontdict, \
           labelpad = labelpad)
plt.ylabel('density', \
           fontdict = label_fontdict, \
           labelpad = labelpad)
plt.title('Average Annual Revenue Distribution of Superhost and Non-superhost', \
          fontdict = label_fontdict, \
          pad = pad);

plt.savefig("figures/Q2.png", bbox_inches = 'tight');
```

**Explanation & Conclusion:** <br>

The obtained result was HK\$ 2633 with a margin error of HK\$1897, which means the lower bound of the estimation did not include zero. In other words, we are 99% confident that at least 50% of super hosts has more average annual revenue than the lower half of the non-super host group. <br>

Therefore, we could say that the title of super host generally brings larger revenue and it worths the extra efforts

### Question 3: Which month is best for visits in terms of availability and pricing?

**Analysis:** <br>
The past availability of listings is not provided by the dataset. We have to estimate it with the assumption that the review number is inversely proportional to the availability. Then, we need to obtain the monthly review numbers since the date of the earliest review so that we can plot out the change over time to observe the seasonality. The month probably having more available listings could thus be spotted out.  <br>

About the pricing, we simply need to obtain the monthly average listing prices in the period of 2021-09-24 - 2022-0924 to find the month that listings generally set lower prices.

```python
# get calendar dataframe for Q3
df_Q3_calendar = df_necessary_column_calendar.copy()
```

```python
# get review dataframe for Q3
df_Q3_review = df_necessary_column_review.copy()
```

```python
# Calculate monthly review number
df_monthly_review_num = df_Q3_review.set_index('date').resample('MS').count()

# Decomposing the time series
decomposition = seasonal_decompose(df_monthly_review_num)

# Get the trend, seasonality and noise
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

### Question 5: What are the impact of the 2019 Hong Kong protests and the COVID-19 pandemic from the Airbnb perspective?

**Analysis**: <br>

Again, we employ the revenues obtained in Q1 as the dependent variable to do the comparison. But, this time, we need to calculate the total revenue in a fixed period. Given that the active period of the 2019 Hong Kong protests was 2019-06-15 - 2020-01-15, we are going to compare the estimated total revenues of the same period in 2018, 2019, and 2020.

#### Q5 calculations

##### prepare listing & review dataframe for Q5 calculations

```python
# obtain from Q1
df_Q5_listing
```

```python
# get review dataframe for Q5
df_Q5_review = df_necessary_column_review.copy()
```

##### calculate total revenues during (2018-06-15 - 2019-01-15), (2019-06-15 - 2020-01-15) & (2020-06-15 - 2021-01-15)

```python
df_review_count_2018 = \
df_Q5_review.where((df_Q5_review['date'] >= '2018-06-15') & (df_Q5_review['date'] <= '2019-01-15'))\
            .groupby(['listing_id'])\
            .count()\
            .rename(columns={'date': 'review_num_2018'})

df_review_count_2019 = \
df_Q5_review.where((df_Q5_review['date'] >= '2019-06-15') & (df_Q5_review['date'] <= '2020-01-15'))\
            .groupby(['listing_id'])\
            .count()\
            .rename(columns={'date': 'review_num_2019'})

df_review_count_2020 = \
df_Q5_review.where((df_Q5_review['date'] >= '2020-06-15') & (df_Q5_review['date'] <= '2021-01-15'))\
            .groupby(['listing_id'])\
            .count()\
            .rename(columns={'date': 'review_num_2020'})
```

```python
# Merge mulitple dataframes, and only get the listings after filtering in Q1
df_list = [df_Q5_listing, df_review_count_2018, df_review_count_2019, df_review_count_2020]

df_merged = reduce(lambda  left, right: pd.merge(left, right, on = ['listing_id'], how = 'left'), df_list)

# Assign 0 to the null elements the calculate the revenue of 2018, 2019, 2020
year = ['2018', '2019', '2020']

for x in year:
    df_merged.at[ df_merged['review_num_{}'.format(x)].isnull(), 'review_num_{}'.format(x) ] = 0
    df_merged['estimated_revenue_{}'.format(x)] = df_merged.loc[:, 'price'] * df_merged.loc[:, 'review_num_{}'.format(x)]

df_merged
```

```python
year = ['2018', '2019', '2020']
column = []

for x in year:
    column.append('estimated_revenue_{}'.format(x))

df_merged[column].sum()
```

#### Q5 results

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (8, 5))

df_merged[column].sum().plot(kind = 'bar', ax = ax)

# ticks formatting
tick_fontsize = 10
ax.set_xticklabels(labels = ['2018-06-15 - 2019-01-15', '2019-06-15 - 2020-01-15', '2020-06-15 - 2021-01-15'], \
                   rotation = 0)
ax.tick_params(axis = 'both', labelsize = tick_fontsize)

# grid line formatting
ax.grid(b = True, axis = 'y', linestyle = '--', linewidth = 1.5)

# labels & title formatting
label_fontdict = \
{'fontsize': 12, \
 'fontweight' : 'semibold'}
labelpad = 20
pad = 30
ax.set_xlabel('periods', \
              fontdict = label_fontdict, \
              labelpad = labelpad) 
ax.set_ylabel('total revenue (HK$)', \
              fontdict = label_fontdict, \
              labelpad = labelpad) 
ax.set_title('Estimated Total Revenue in Years of 2018, 2019 and 2020', \
             fontdict = label_fontdict, \
             pad = pad);

plt.savefig("figures/Q5.png", bbox_inches = 'tight');
```

**Explanation & Conclusion:** <br>

Overall, a steady declining trend can be observed from the figure. The first period, 2018â€“06â€“15â€“2019â€“01â€“15, had the greatest revenue (~HK$ 10,950,000) among the three of them. The second period was approximately three-quarters of the first one. And the last period approximately halved compared to the second one. <br>

Therefore, we could conclude that there was a decline in total revenue when the social movements took place and a more significant drop was observed when the COVID-19 outbreak happened.
